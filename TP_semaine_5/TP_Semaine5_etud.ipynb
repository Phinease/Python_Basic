## Introduction

Le module `matplotlib` permet de tracer des fonctions. Par exemple, pour tracer $x \mapsto \sqrt{x}$, le code minimal est le suivant :

```python
# Définition des valeurs pour l'abscisse
x = np.linspace(0,10,5)

# Création d'une figure contenant un graphe
fig, ax = plt.subplots() 
# Ajout de la courbe
ax.plot(x,np.sqrt(x))
```

La courbe consiste à "relier" les points de coordonnées $(x[i],y[i])$ ; sur l'exemple ci-dessous, il n'y pas assez de points donc la courbe n'est pas assez arrondie. Si l'on prend plus de points, on reconnaît mieux la fonction racine carrée :

```python
# Définition des valeurs pour l'abscisse
x = np.linspace(0,10,150)
print(np.cos(x)-x**3)

# Création d'une figure contenant un graphe
fig, ax = plt.subplots() 
# Ajout de la courbe
ax.plot(x,np.sqrt(x),label="racine carrée")# On trace une première courbe, le label sert pour la légende
ax.plot(x,x,label="identité")# On trace une seconde courbe.
ax.legend() # Obligatoire pour afficher les légendes
# Ajout des titres des axes :
ax.set_xlabel("x") # titre de l'axe des  abscisses
ax.set_ylabel("x") # titre de l'axe des ordonnées
ax.set_title("mes des courbes") # titre du graphe
```

Il est possible d'afficher plusieurs graphes (côte-à-côte, ou sur plusieurs lignes). Tester le code suivant :

```python
def f(x):
    return np.sin(x)
    
def test_plot():
    s =0.
    e = 4.*np.pi
    x = np.linspace(s,e, 150)
    y = f(x)
    fig, ax = plt.subplots(2,1,figsize=(10,5))
    ax[0].plot(x, y,label="ma courbe")
    ax[0].set_title(r'$y=\sin(x)$')
    ax[0].set_xlabel("x")
    ax[0].set_ylabel("sinus")
    ax[0].legend()
    
    ax[1].plot(x, np.cos(x),label="ma courbe")
    ax[1].set_title(r'$y=\cos(x)$')
    ax[1].set_xlabel("x")
    ax[1].set_ylabel("cosinus")
    ax[1].legend()
    
    fig.suptitle("Fonctions trigonométriques élémentaires")
    #fig.show() # Inutile dans les notebooks uniquement.

test_plot()
```

* sur le graphe de gauche, $t \mapsto \sin(\omega_1 t +\pi)$, $t \mapsto \sin(\omega_2 t)$ puis leur somme,
* sur le graphe de droite, $t \mapsto 0.8 \sin(\omega_1 t +\pi)$, $t \mapsto 2 \sin(\omega_2 t)$ puis leur somme,

avec $\omega_1 = 1$, $\omega_2 = 1.1$ et pour $t \in [0,30\pi]$.

```python
from numpy import *

def plot_2 ():
    w1 = 1.
    w2 = 1.1
    t = linspace(0,30*pi,200)
    y1 = sin(w1*t+pi)
    y2 = sin(w2*t)
    y3 = 0.8*y1
    y4 = 2*y2
    fig, ax = plt.subplots(2,1,figsize=(10,5))
    ax[0].plot(t, y1, label="f1")
    ax[0].plot(t, y2, label="f2")
    ax[0].plot(t, y1+y2, label="somme")
    ax[0].set_title(r'$y = \exp(whatever)$')
    ax[0].set_xlabel("x")
    ax[0].set_ylabel("y")
    ax[0].legend()
    
    ax[1].plot(t, y3, label="f1")
    ax[1].plot(t, y4, label="f2")
    ax[1].plot(t, y3+y4, label="somme")
    ax[1].set_title(r'$y=\sin(x)$')
    ax[1].set_xlabel("x")
    ax[1].set_ylabel("y")
    ax[1].legend()
    
plot_2()
```

Pour tracer une courbe paramétrée $x(t),y(t)$, il suffit de définir correctement les tableaux `x` et `y` de façon à ce que `x[i]` et `y[i]` soient les coordonnées de la courbe à l'instant `t[i]`.

### Question
1. Pour $t\in [0;2\pi]$, tracer la courbe paramétrée :
\begin{equation}
\left\{
\begin{aligned}
x(t) &= \sin^3 (t) \\
y(t) &= \cos(t) - \cos^4 (t)
\end{aligned}
\right.
\end{equation}
2. Tracer la courbe de Lissajou suivante :
\begin{equation}
\left\{
\begin{aligned}
x(t) &= \sin(5t) \\
y(t) &= \sin(3t +\pi)
\end{aligned}
\right.
\end{equation}




```python
def plot_3 ():
    w1 = 1.
    w2 = 1.1
    t = linspace(0,2*pi,200)
    y1 = pow(sin(t),3)
    y2 = cos(t)-pow(cos(t),4)
    y3 = sin(5*t)
    y4 = sin(3*t+pi)
    fig, ax = plt.subplots(2,1,figsize=(10,5))
    ax[0].plot(t, y1, label="x(t)")
    ax[0].plot(t, y2, label="y(t)")
    ax[0].set_title("courbe paramétrée")
    ax[0].set_xlabel("x")
    ax[0].set_ylabel("")
    ax[0].legend()
    
    ax[1].plot(t, y3, label="x")
    ax[1].plot(t, y4, label="y")
    ax[1].set_title("courbe de Lissajou")
    ax[1].set_xlabel("t")
    ax[1].set_ylabel("")
    ax[1].legend()
    
    
plot_3()
```

## Paramétrisation polaire
Cette paramétrisation consiste à repérer un point en fonction d'un angle (le zéro correspondant à l'axe des abscisse) et d'un rayon (distance à l'origine de l'axe), par exemple l'angle $\theta$ pour un cercle $x = rcos(\theta)$, $y = rsin(\theta)$.
Matplotlib fournit _nativement_ des outils pour tracer des courbes en coordonnées polaires.

Pour pouvoir définir une courbe à l'aide des coordonnées polaires, il faut utiliser la projection polaire : `projection='polar'`. En voici deux exemples :

```python
t = np.linspace(0,2.*np.pi,200)

# Exemple 1 : uniquement des polaires
fig, ax = plt.subplots(1,2,figsize=(10,2),subplot_kw=dict(projection='polar'))
ax[0].plot(t,np.ones(t.shape))
ax[1].plot(np.sin(t),np.cos(t))
fig.subplots_adjust(top=0.8,wspace=0.) # Pour ajuster l'espacement (wspace pour horizontal, hspace pour vertical)
fig.suptitle("Uniquement des courbes polaires")

# Exemple 2 : mélange
fig2 = plt.figure()
ax0 = fig2.add_subplot(121, projection='polar')
ax1 = fig2.add_subplot(122, aspect='equal')
ax0.plot(t,np.ones(t.shape))
ax1.plot(np.sin(t),np.cos(t))
fig2.subplots_adjust(top=1.1,wspace=0.5) # Pour ajuster l'espacement (wspace pour horizontal, hspace pour vertical)
fig2.suptitle('Une courbe en polaire et une "classique"')
[]
```

````
import numpy.random as rd

def _matrix_random_anyway(n) :
  return rd.random([n, n])

m = _matrix_random_anyway(5);
print(m)

def matrix_random_symetrique(n) :
  m = _matrix_random_anyway(n)
  return 0.5*(m.transpose()+m)
````
